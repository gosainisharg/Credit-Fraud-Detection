# Dealing with Unbalanced Data
## We will use undersampling
Build a sample dataset from original dataset having similar distribution of legit and fraudulent transactions
Randomly pick 492 legit transaction and all 492 fraud transactions

```python
#pick 492 legit transactions using random sampling
legit_sample = legit_tr.sample(n=492)
```

```python
#concat legit sample and fraud transac
new_crd_data = pd.concat([legit_sample,fraud_tr], axis = 0) #we need to mention axis = 0 this will add row wise, we dont want column wise
```

```python
new_crd_data.head()
```

```python
new_crd_data.tail()
```

```python
new_crd_data.describe()
```

```python
new_crd_data['Class'].info()
```

```python
new_crd_data['Class'].value_counts()
```

```python
new_crd_data.groupby('Class').mean()
```

```python
#nature of the dataset has not changed
#this means we got a good sample, if the means were changed drastically it would mean a bad sample
```

# Splitting dataset into features and targets
X -> Features
Y -> Targets

```python
X = new_crd_data.drop(columns = 'Class',axis = 1) #drop Class column
Y = new_crd_data['Class']
```

```python
print(X)
print(Y)
```

# Splitting data into training and testing data
##We will use train_test_split(),
##20% of data will be testing data
###stratify insures even distribution of 0 and 1 in test and train data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Model Training
## Logistic Regression

```python
model = LogisticRegression()
```

```python
#training model using training data
model.fit(X_train,Y_train)
```

```python
#very less data points, training can take long time if more data points
```

# Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data:',training_data_accuracy)
```

```python
#good accuracy score on training data
#but need to check on test data
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on testing data:',test_data_accuracy)
```

If accuracy scores are different for training and test data then we need to check for overfitting or underfitting, but we are fine in this case.
